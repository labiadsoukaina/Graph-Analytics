```python
data = pd.read_csv("airports_distance")
sorted_data = data.sort_values(by=['distance'])
sorted_df = pd.DataFrame(sorted_data)
```

```python
sorted_df.head()
```

```python
airports_faible_distance = sorted_df[sorted_df['distance'] <150]
airports_faible_distance
```

```python
airports_faible_distance_source = airports_faible_distance['source'].tolist()
airports_faible_distance_target = airports_faible_distance['target'].tolist()

airports_faible_distance_list = airports_faible_distance_source + airports_faible_distance_target
airports_faible_distance_list = list(dict.fromkeys(airports_faible_distance_list))

print("Liste des aéreports à supprimer:")
print(len(airports_faible_distance_list))
for airport in airports_faible_distance_list:
    print("-",airport)
```

```python
airports = pd.read_csv("airports.csv")
airports_df = pd.DataFrame(airports)
airports_df.head()
```

```python
def return_location(airport_name):
    location = []
    airport = airports_df[airports_df['Name'] == airport_name]
    lat = airport.iloc[0]['Latitude']
    long = airport.iloc[0]['Longitude']
    location.append(lat)
    location.append(long)
    return location
```

```python
import folium
def show_me_my_airport(my_airport):
    loc = return_location(my_airport)
    my_map = folium.Map(location=loc)
    my_map.save(my_airport+".html")
    return my_map
```

```python
for airport in airports_faible_distance_list:
    show_me_my_airport(airport)
```
